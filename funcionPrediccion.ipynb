```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
import cv2
```

```python
longitud_m1_m3, altura_m1_m3 = 150, 150
longitud_m2, altura_m2 = 200, 200
pred = ''
```

```python
modelo_m1 = './modelo/modelo1_project.h5'
pesos_modelo_m1 = './modelo/pesos1_project.h5'
modelo_m2 = './modelo/modelo2_project.h5'
pesos_modelo_m2 = './modelo/pesos2_project.h5'
modelo_m3 = './modelo/modelo3_project.h5'
pesos_modelo_m3 = './modelo/pesos3_project.h5'
```

```python
cnn_m1 = load_model(modelo_m1)
cnn_m1.load_weights(pesos_modelo_m1)

cnn_m2 = load_model(modelo_m2)
cnn_m2.load_weights(pesos_modelo_m2)

cnn_m3 = load_model(modelo_m3)
cnn_m3.load_weights(pesos_modelo_m3)
```

```python
plt.figure(figsize=(10, 10))
```

```python
def predict_m1(file):
  #x = load_img(file, target_size=(longitud_m1_m3, altura_m1_m3))
  x = plt.imread(file)
  x= cv2.resize(x,(longitud_m1_m3,altura_m1_m3))
  x= cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
  plt.subplot(3, 1, 1)
  plt.title('Moledo 1')
  plt.imshow(x)
  #plt.show()
  x = img_to_array(x)
  x = np.expand_dims(x, axis=0)
  print(x.shape)
  array = cnn_m1.predict(x)
  result = array[0]
  answer = np.argmax(result)
  if answer == 0:
    pred = 'nano'
    print("pred: Nano")
  elif answer == 1:
    pred = 'Uno'
    print("pred: Uno")

  plt.axis('off')
  plt.title('Modelo 1, Se supone que es: {}'.format(pred))
  return result
```

```python
def predict_m2(file):
  x = load_img(file, target_size=(longitud_m2, altura_m2))
  plt.subplot(3, 1, 2)
  plt.imshow(x)
  x = img_to_array(x)
  x = np.expand_dims(x, axis=0)
  print(x.shape)
  array = cnn_m2.predict(x)
  result = array[0]
  answer = np.argmax(result)
  if answer == 0:
    pred = 'nano'
    print("pred: Nano")
  elif answer == 1:
    pred = 'Uno'
    print("pred: Uno")

  plt.axis('off')
  plt.title('Modelo 2, Se supone que es: {}'.format(pred))
  return result
```

```python
def predict_m3(file):
  x = load_img(file, target_size=(longitud_m1_m3, altura_m1_m3))
  plt.subplot(3, 1, 3)
  plt.imshow(x)
  x = img_to_array(x)
  x = np.expand_dims(x, axis=0)
  print(x.shape)
  array = cnn_m3.predict(x)
  result = array[0]
  answer = np.argmax(result)
  if answer == 0:
    pred = 'nano'
    print("pred: Nano")
  elif answer == 1:
    pred = 'Uno'
    print("pred: Uno")
  plt.axis('off')
  plt.title('Modelo 3, Se supone que es: {}'.format(pred))
  #plt.title('Training and Validation Loss')
  return result
```
